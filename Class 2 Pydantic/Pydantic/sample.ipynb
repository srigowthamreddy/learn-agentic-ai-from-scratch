```python
def register_user(name, email, age):
    birth_year = 2026 - age
    print(f"Registering user : Name: {name}, Email: ({email}), Birth Year: {birth_year}")
```

```python
good_user_data = {
    "name" : "Gowtham",
    "email" : "gowtham@example.com",
    "age" : 23
}
register_user(**good_user_data)
```
